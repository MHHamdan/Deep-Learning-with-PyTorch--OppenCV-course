##  <font color='green'>Step 3: Build C++ Code</font>

We have provided a `CMakeLists.txt` file in the sample code folder. You can use the following commands to build the C++ code (`sample.cpp` in our case).


**NOTE** that you need to provide the path where you have stored the `libtorch` folder in the `CMakeLists.txt`.


Let's have a look inside the `sample.cpp` and `CMakelists.txt`.

**`sample.cpp`**

```c++
#include <torch/torch.h>
#include <iostream>

using namespace std;

int main() {

  torch::Tensor tensor = torch::randn({3,3});
  cout << "The random matrix is:" << endl << tensor << endl;


  // Initialize the device to CPU
  torch::DeviceType device = torch::kCPU;
  // If CUDA is available,run on GPU
  if (torch::cuda::is_available())
      device = torch::kCUDA;
  cout << "Running on: "
            << (device == torch::kCUDA ? "GPU" : "CPU") << endl;

  return 0;

}
```

Even if we do not understand the `LibTorch` syntax, we can say that the `CPP` code is written to create and print a random tensor and check if the machine has GPU. 


**`CMakelists.txt`** Please note that we have placed the path to the downloaded folder of `Libtorch` in `find_package`


```bash
cmake_minimum_required(VERSION 3.0 FATAL_ERROR)
project(sample-code)

find_package(Torch REQUIRED PATHS "../libtorch" )

add_executable(sample sample.cpp)
target_link_libraries(sample "${TORCH_LIBRARIES}")
set_property(TARGET sample PROPERTY CXX_STANDARD 14)


```

**Let's see how to build and run.**

###  <font color='green'>Ubuntu/MacOS Users</font>

```bash
cd libtorch-sample-code-master
mkdir build
cd build
cmake -DCMAKE_PREFIX_PATH=<Specify LIBTORCH_PATH here> .. 
# ( e.g. cmake -DCMAKE_PREFIX_PATH=$PWD/../libtorch .. )
cmake --build . --config Release
cd ..
./build/sample
```

###  <font color='green'>Windows Users</font>

```bash
cd libtorch-sample-code-master
mkdir build
cd build
cmake -DCMAKE_PREFIX_PATH=<Specify FULL_PATH_TO_LIBTORCH_PATH here> .. 
# ( e.g. cmake -DCMAKE_PREFIX_PATH= C:\Users\Administrator\Downloads\libtorch .. )
cmake --build . --config Release
cd ..
.\build\Release\sample.exe
```
